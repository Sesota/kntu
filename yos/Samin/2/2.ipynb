## Summation of sequences

$$
\large
a=\{a_0,a_1,a_2,a_3,...,a_{n-1}\}
$$

$$
\large
S=\sum_{i=0}^{n}a_i
$$

$$
\large
S=a_0+a_1+a_2+...+a_{n-1}
$$

$$
S=a_0+a_1+a_2+...+a_{n-1} \\
=(d*0+a_0)+(d*1+a_0)+(d*2+a_0)+...+(d*(n-1)+a_0) \\
=d*(0+1+2+...+(n-1))+n*a_0 \\
=d\frac{n(n-1)}{2}+n*a_0 \\
\Large
=\frac{n}{2}(d(n-1)+2a_0)
$$

### Example
$\Large 2+5+8+...+56=?$

$
S=\frac{n}{2}(d(n-1)+2a_0) \\
d=3 \\
a_0=2 \\
\large n=? \\
$

$
a_{n-1}=d*(n-1)+a_0 \\
\rightarrow 56=3*(n-1)+2 \\
\rightarrow \frac{54}{3}+1=n \\
\large \rightarrow n=19 \\
$

$
S=\frac{n}{2}(d(n-1)+2a_0) \\
d=3 \\
a_0=2 \\
n=18 \\
\rightarrow S=\frac{19}{2}(3*(19-1)+2*2) \\
\rightarrow S=\frac{19}{2}(3*18+4)=9(54+4)=\frac{19*58}{2} \\
\Large \rightarrow S=19*29=551 \\
$

### Example
$\Large 4+5+6+...+x=814 \\
\Large \rightarrow x=?$

$
S=\frac{n}{2}(d(n-1)+2a_0) \\
d=1 \\
a_0=4 \\
\rightarrow 814=\frac{n}{2}(1*(n-1)+2*4) \\
\rightarrow 814=\frac{n}{2}(n-7) \\
\rightarrow 2*814=n(n-7) \\
$

$
2*814=4*407=44*37 \\
\rightarrow n=37 \\
\Large \rightarrow a_{n-1}=d*(n-1)+a_0=36+4=40 \\
$
