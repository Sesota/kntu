```python
import matplotlib.pyplot as plt
import timeit
from ipywidgets import interact
```

```python
def bubble_sort(array):
    for j in range(len(array)-1):
        for i in range(len(array)-j-1):
            if array[i]>array[i+1]:
                array[i],array[i+1] = array[i+1],array[i]
                
    return array
```

```python
arrays = [list(reversed(range(i))) for i in range(30)]
res = []
@interact(n=(1,10))
def prof(n):
    for j in range(5,15):
        times = []
        for i in arrays:
            r = timeit.repeat(lambda: bubble_sort(i), number=n, repeat=j)
            times.append(min(r))
        res.append(times)

    plt.subplot(1, 2, 1)
    for i,v in enumerate(res):
        plt.plot(v, label=f"{i}")

    plt.subplot(1, 2, 2)
    for i,v in enumerate(res):
        plt.plot(v, label=f"{i}")
    plt.legend()
```

```python
import matplotlib.pyplot as plt
plt.plot(times)
plt.plot(times)
```
